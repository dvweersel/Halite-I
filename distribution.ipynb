```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.set()
```

```python
all_files = os.listdir('training_data')

rows = []
for idx, f in enumerate(all_files):
    avg_halite = float(f.split("-")[0])
    players = int(f.split("-")[1])
    gathered = float(f.split("-")[2])
    
    row = {'avg_halite' : avg_halite,
          'players' : players,
          'gathered': gathered}
    rows.append(row)

df = pd.DataFrame(rows, columns=['avg_halite', 'players', 'gathered'])
```

```python
df['halite_pp'] = df['avg_halite']/df['players']
```

```python
sns.scatterplot(x='avg_halite', y='gathered', style='players', data=df)
```

```python
sns.scatterplot(x='avg_halite', y='gathered', data=df)
```

```python
sns.pairplot(df)
```
