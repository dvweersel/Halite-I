```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.set()
```

```python
all_files = os.listdir('training_data')
rows = []

for f in all_files:
	avg_halite = float(f.split("-")[0])
	players = int(f.split("-")[1])
	gathered = float(f.split("-")[2])
	file = f

	row = {'avg_halite' : avg_halite,
		'players' : players,
		'gathered': gathered,
		'file' : f}

	rows.append(row)

df = pd.DataFrame(rows, columns=['avg_halite', 'players', 'gathered', 'file'])

pd.DataFrame.sort_index

df.sort_values('gathered', ascending=False).groupby('avg_halite', sort=False).head(5).sort_values('avg_halite').plot.scatter(x='avg_halite', y='gathered')
```

```python
df.sort_values('gathered', ascending=True).groupby('avg_halite', sort=False).head(2).sort_values('avg_halite').plot.scatter(x='avg_halite', y='gathered')
```

```python
games = df.sort_values('gathered', ascending=False).groupby('avg_halite', sort=False).head(5)

for f in games.head(5).values:
    print(f[3])

training_file_names = []
for f in games.values:
	training_file_names.append(os.path.join('taining_data', f[3]))
```
